```python
from nltk import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from string import punctuation as znaki
from collections import Counter
from nltk.corpus import stopwords
import matplotlib.pyplot as mll
lem = WordNetLemmatizer()
with open('C:\Рабочий стол\Дз\Coding\Harry_Potter_and_the_Sorcerers_Stone.txt', 'r', encoding='utf-8') as hp:
    text = hp.read()   
words = [word.lower().strip(znaki+"«»—") for word in word_tokenize(text)] #Приводит текст к нижнему регистру и убирает пунктуацию
words = list(filter(None, words)) #Удаляет пустые строки
lemma = [lem.lemmatize(w) for w in words]
with open('C:\Рабочий стол\Дз\Coding\Harry_Potter_and_the_Sorcerers_Stone1.txt', 'w', encoding='utf-8') as hp1:
    hp1.write(' '.join(lemma))    
def rel(words):
    absolut = list(Counter(words).items()) #Вычисляет абсолютную частоту
    absolut.sort(reverse=True, key=lambda x: x[1]) #Сортирует словарь по частоте
    count = len(words) #Вычисляет количество слов
    return [(w, c / count) for w, c in absolut] #Получает относительную чатсоту
otnosit=rel(lemma)
mll.plot(range(1, len(otnosit) + 1),
    [f for w, f in otnosit])
mll.show() #График - проверка законца Ципфа
stopword = stopwords.words('english')
notstopword = [(w, f) for w, f in otnosit if w not in stopword]
samyi = notstopword[:30] #Выбирает 30 самых частотных слов
mll.bar(range(30), [f for w, f in samyi], edgecolor='pink')
mll.xticks(range(30), [w for w, f in samyi], rotation=90)
mll.show() #barplot для 30 самых частотных слов
```
