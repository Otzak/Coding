## Задача 1 

Скачайте текст ["Литературных анекдотов"](https://github.com/ancatmara/python-for-dh/blob/master/Classes/9-10/literary_anecdotes.txt). 
Напишите функцию, которая будет читать файл, лемматизировать текст с помощью `pymystem3` и записывать результат в новый файл. 
У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
def reader(ishodny, done): #заявляет функцию reader
    text = open(ishodny, 'r', encoding='utf-8') #открывает исходный файл
    from pymystem3 import Mystem
    m=Mystem(mystem_bin="C:\Рабочий стол\Дз\Coding\mystem.exe")
    everything=''.join(m.lemmatize(text.read())) #лемматизирует текст
    text.close()
    with open(done, 'w', encoding='utf-8') as newtext:
        newtext.write(everything) #записывает полученный текст в новый файл
    
reader('C:\Рабочий стол\Дз\Coding\literary_anecdotes.txt', 'C:\Рабочий стол\Дз\Coding\literary_anecdotes1.txt')
```

## Задача 2
Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
with open(r'C:\Рабочий стол\Дз\Coding\literary_anecdotes1.txt', 'r', encoding='utf-8') as l: #открывает файл, созданный в предыдущей программе
    slova = l.read().split() #разделяет этот текст на слова
with open(r'C:\Рабочий стол\Дз\Coding\rus_stopwords.txt', 'r', encoding='utf-8') as rsw: #открывает файл rus_stopwords со стоп-словами
    stop_slova = [sw.rstrip() for sw in rsw.readlines()] #убирает из списка стоп-слов переносы строки
itogo = [clova for clova in slova if clova not in stop_slova] #убирает из текста стоп-слова
slovar = {}
for clova in itogo:
    if clova in slovar:
        slovar[clova] += 1
    else:
        slovar[clova] = 1 #создает словарь встречающихся в тексте слов и подсчитывает их количество
ipm = {} #создает словарь ipm
for key, value in slovar.items():
    ipm[key] = value / len(slova) * 1000000 #Вычисляет ipm
sorted_counts = sorted(ipm.items(), key=lambda x: x[1], reverse=True) #сортирует (способом lambda) ipm словарь по частотности
print(sorted_counts[:20]) #выводит 20 самых частотных слов
```

## Задача 3
Сделайте полный морфологический разбор исходного текста. Напишите регулярное выражение, которое будет извлекать из тега только часть речи.
Пройдитесь циклом по списку с разборами, который выдал `pymystem3`, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение).
Посчитайте абсолютную частоту для всех частей речи, а затем относительнную частоту (абсолютная частота / длина текста в словах).

```python
import re
chasti = re.compile('[A-Z]+') #ищет части речи с помощью регулярного выражения (так как части речи обозначаются заглавными латинскими буквами)
with open('C:\Рабочий стол\Дз\Coding\literary_anecdotes.txt', 'r', encoding='utf-8') as slova:
    bylo = slova.read()
from pymystem3 import Mystem
m=Mystem(mystem_bin="C:\Рабочий стол\Дз\Coding\mystem.exe")
analiz = m.analyze(bylo) #делает морфологический анализ
slovar1 = {}
for slovar2 in analiz: #цикл для списка словарей
    for key, value in slovar2.items(): #цикл для словарей
        try:
            slovo = slovar2['text']
            grammar = slovar2['analysis'][0]['gr']
            slovar1[slovo] = chasti.match(grammar).group(0)
        except (KeyError, IndexError) as e: #избавляет от ошибок типа KeyError и IndexError
            pass
print(slovar1)
slovar3 = slovar1.values() #список со значениями полученного словаря
from collections import Counter
kolichestvo = Counter(slovar3) #считает количество значений в полученном словаре
print(kolichestvo)
slovar4 = {} #создает словарь, нужный для вывода относительной частотности
for key, value in kolichestvo.items():
    chast = value / len(analiz) #считает относительную частотность
    slovar4[key] = chast #Соотносит относительную частотность с элементами в словаре
print(sorted(slovar4.items(), key=lambda x: x[1], reverse=True)) #Выводит относительную частотность
```
